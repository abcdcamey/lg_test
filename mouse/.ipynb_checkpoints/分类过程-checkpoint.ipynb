# 分类思路

```
1、把分割出的图片大小统一resize到32x32的大小，然后使用AutoEncoder进行图片的特征提取，提取后的特征维度为:4x4x3,实现降维，维度由1024->48。
2、每张图片，使用降维后的48个维度特征，应用k-means算法进行分类，类别数量参数设置为3
3、根据k-means分类出的结果，大致可以把细胞形态分为:饱满圆形细胞、偏长条形细胞、其他形态细胞
```

## 导入依赖包

````python
import torch
import torch.nn as nn
import random
import cv2 as cv
import torch.nn.functional as F
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
````

## 读取分割后的细胞图片，并resize到32x32的大小

````python
def resize_img(img):
    (y,x) = img.shape
    max_edge = max(x,y)
    resize_img = cv.resize(img,(int(x/max_edge*32),int(y/max_edge*32)),interpolation = cv.INTER_NEAREST)
    if resize_img.shape[0]==32:
        gap = 32-resize_img.shape[1]
        resize_img = cv.copyMakeBorder(resize_img,0,0,gap//2,gap-gap//2,cv.BORDER_CONSTANT)
    else:
        gap = 32-resize_img.shape[0]
        resize_img = cv.copyMakeBorder(resize_img,gap//2,gap-gap//2,0,0,cv.BORDER_CONSTANT)
    return resize_img



# 读取图片数据
image_list = []
image_paths = glob.glob("images/cell/*_rotate.tif")
for path in image_paths:
    
    gray_img = cv.imread(path,cv.IMREAD_GRAYSCALE)
    resized_img = resize_img(gray_img)
    
    image_list.append(resized_img)
train_data = np.array(image_list)
    
````

## AutoEncoder模型构造和训练

````python
class AutoEncoder(nn.Module):
    def __init__(self):
        super(AutoEncoder, self).__init__()
        
        self.conv1 = nn.Conv2d(1, 8, 3, padding=1)
        self.conv2 = nn.Conv2d(8, 5, 3, padding=1)
        self.conv3 = nn.Conv2d(5, 3, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.t_conv1 = nn.ConvTranspose2d(3, 5, 2, stride=2)
        self.t_conv2 = nn.ConvTranspose2d(5, 8, 2, stride=2)
        self.t_conv3 = nn.ConvTranspose2d(8, 1, 2, stride=2)
        
        
    def forward(self,x):
        ## encode
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)  
        x = F.relu(self.conv3(x))
        x = self.pool(x)
        ## decode
        x = F.relu(self.t_conv1(x))
        x = F.relu(self.t_conv2(x))
        # output layer (with sigmoid for scaling from 0 to 1)
        x = F.sigmoid(self.t_conv3(x))
        
        return x
    
    #只做encoder提取压缩后特征
    def encode(self,x):
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = F.relu(self.conv3(x))
        x = self.pool(x)
        return x
````

````python
model = AutoEncoder()
````

````python
criterion = torch.nn.MSELoss()
lr= 0.0005
optim = torch.optim.Adam(model.parameters(), lr=lr)
batch_size = 32
epoch_num = 3500
````

````python
for epoch in tqdm(range(epoch_num)):
    cnt=0
    train_loss = 0.0
    for i in range(0,len(train_data),batch_size):
        index_start = i
        batch_images = train_data[i:i+batch_size]/255
        batch_images = torch.Tensor(batch_images)
        batch_images = torch.unsqueeze(batch_images,1)
        
        optim.zero_grad()
        outputs = model(batch_images)
        loss = criterion(outputs, batch_images)
        loss.backward()
        optim.step()
        train_loss += loss.item()*batch_images.size(0)
        cnt+=1
    train_loss = train_loss/cnt
    if epoch%100==0:
        print('Epoch: {} \tTraining Loss: {:.6f}'.format(
            epoch, 
            train_loss
            ))
````

## 使用K-means进行聚类

````python
n_clusters = 3
km = KMeans(n_clusters=n_clusters)
km.fit(batch_images_encoded.data)
# 查看 clusters 的分布结果
plt.figure(figsize=(20,5))
cluster_elements = [(km.labels_==i).sum() for i in range(n_clusters)]
plt.bar(range(n_clusters), cluster_elements, width=0.5)
````
